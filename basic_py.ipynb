```
x = [1, 2, 5];
y = [2, 3, 7];

title = "Gráfico de barras em Python!";
eixoX = "Eixo X";
eixoY = "Eixo Y";

## Titulo do Gráfico
plt.title("Gráfico de linhas em Python!");

## Legenda dos eixos
plt.xlabel("Eixo X");
plt.ylabel("Eixo Y");

## plt.plot = grafico de linhas
plt.plot(x, y);
plt.show();
```

```
x = [1, 2, 3, 4, 5];
y = [2, 3, 7, 1, 0];

## Titulo do Gráfico
plt.title(title);

## Legenda dos eixos
plt.xlabel(eixoX);
plt.ylabel(eixoY);

## plt.bar = grafico de barras
plt.bar(x, y);
plt.show();
```

```
x1 = [1, 3, 5, 7, 9];
y1 = [2, 3, 7, 1, 0];

x2 = [2, 4, 6, 8, 10];
y2 = [5, 1, 3, 7, 4];

## Titulo do Gráfico
plt.title(title);

## Legenda dos eixos
plt.xlabel(eixoX);
plt.ylabel(eixoY);

## plt.bar = grafico de barras
plt.bar(x1, y1, label = "Grupo 1");
plt.bar(x2, y2, label = "Grupo 2");

## plt.legend = coloca legenda no grafico
plt.legend();
plt.show();
```

```
x1 = [1, 3, 5, 7, 9];
y1 = [2, 3, 7, 1, 0];

x2 = [2, 4, 6, 8, 10];
y2 = [5, 1, 3, 7, 4];

title1 = "Gráfico Scatterplot em Python!";

## Titulo do Gráfico
plt.title(title1);

## Legenda dos eixos
plt.xlabel(eixoX);
plt.ylabel(eixoY);

## plt.scatter = grafico scatterplot
## atributo color altera a cor dos pontos no grafico
plt.scatter(x1, y1, label = "Grupo 1", color = "red");
plt.scatter(x2, y2, label = "Grupo 2");

## grafico de linhas dentro do grafico scatterplot
plt.plot(x1, y1, color = "orange");
plt.plot(x2, y2);

## plt.legend = coloca legenda no grafico
plt.legend();
plt.show();
```

```
x1 = [1, 3, 5, 7, 9];
y1 = [2, 3, 7, 1, 0];

x2 = [2, 4, 6, 8, 10];
y2 = [5, 1, 3, 7, 4];

title1 = "Gráfico Scatterplot em Python!";

## Titulo do Gráfico
plt.title(title1);

## Legenda dos eixos
plt.xlabel(eixoX);
plt.ylabel(eixoY);

## plt.scatter = grafico scatterplot
## atributo color altera a cor dos pontos no grafico
## marker = altera o marcador no grafico
## s = altera o tamanho do marcador
## linestyle = altera o estilo da linha
plt.scatter(x1, y1, label = "Grupo 1", color = "red", marker = "^", s = 150);
plt.scatter(x2, y2, label = "Grupo 2", color = "green", marker = "X", s = 150);

## grafico de linhas dentro do grafico scatterplot
plt.plot(x1, y1, color = "orange", linestyle = ":");
plt.plot(x2, y2);

## plt.legend = coloca legenda no grafico
plt.legend();
plt.show();
```

```
## Estudo de caso - Crescimento da população Brasileira (1980 a 2016)
## Informações do DataSUS

## Montar o drive dentro do Google Colab (Executar somente uma vez. Após isso, o código deve ser comentado)
# from google.colab import drive

# drive.mount('/content/drive');

import matplotlib.pyplot as plt
import numpy as np

dados = open('/content/drive/My Drive/populacao_brasileira.csv').readlines();

x = [];
y = [];

for i in range(len(dados)):
  if i != 0:
    linha = dados[i].split(';');
    x.append(int(linha[0]));
    y.append(int(linha[1]));

title = 'Crescimento da população brasileira';
xlabel = 'Ano';
ylabel = 'População x 100.000.000';

plt.title(title);
plt.xlabel(xlabel);
plt.ylabel(ylabel);

plt.scatter(x, y, color = 'blue', marker = '.', s = 100);
plt.plot(x, y, color = 'black');

## plt.grid = coloca as linhas da grade no gráfico
plt.grid();
plt.show();
```

```
## Boxplot - Diagrama de caixa

import matplotlib.pyplot as plt
import random

vetor = [];

for i in range(10):
  num_aleatorio = random.randint(0, 5);
  vetor.append(num_aleatorio);

plt.title('Boxplot!')
plt.boxplot(vetor);
plt.show();
```
