```
x1 = [1, 3, 5, 7, 9];
y1 = [2, 3, 7, 1, 0];

x2 = [2, 4, 6, 8, 10];
y2 = [5, 1, 3, 7, 4];

title1 = "Gr√°fico Scatterplot em Python!";

## Titulo do Gr√°fico
plt.title(title1);

## Legenda dos eixos
plt.xlabel(eixoX);
plt.ylabel(eixoY);

## plt.scatter = grafico scatterplot
## atributo color altera a cor dos pontos no grafico
plt.scatter(x1, y1, label = "Grupo 1", color = "red");
plt.scatter(x2, y2, label = "Grupo 2");

## grafico de linhas dentro do grafico scatterplot
plt.plot(x1, y1, color = "orange");
plt.plot(x2, y2);

## plt.legend = coloca legenda no grafico
plt.legend();
plt.grid();
plt.show();
```

```
x1 = [1, 3, 5, 7, 9];
y1 = [2, 3, 7, 1, 0];

x2 = [2, 4, 6, 8, 10];
y2 = [5, 1, 3, 7, 4];

title1 = "Gr√°fico Scatterplot em Python!";

## Titulo do Gr√°fico
plt.title(title1);

## Legenda dos eixos
plt.xlabel(eixoX);
plt.ylabel(eixoY);

## plt.scatter = grafico scatterplot
## atributo color altera a cor dos pontos no grafico
## marker = altera o marcador no grafico
## s = altera o tamanho do marcador
## linestyle = altera o estilo da linha
plt.scatter(x1, y1, label = "Grupo 1", color = "red", marker = "^", s = 150);
plt.scatter(x2, y2, label = "Grupo 2", color = "green", marker = "X", s = 150);

## grafico de linhas dentro do grafico scatterplot
plt.plot(x1, y1, color = "orange", linestyle = ":");
plt.plot(x2, y2);
plt.grid();

## plt.legend = coloca legenda no grafico
plt.legend();
plt.show();
```

```
## O ponto de acesso ao drive ja foi montado anteriormente.
## Se nao tiver feito o passo anterior para realizar a montagem, ter√° que fazer aqui.

## DNA Bact√©ria
dados_entrada_bacteria = open('/content/drive/My Drive/Data Scientist/bacteria.fasta').read().replace('\n', '');
# dados_entrada_human = open('/content/drive/My Drive/Data Scientist/human.fasta').read().replace('\n', '');

dados_saida_bacteria = open('/content/drive/My Drive/Data Scientist/bacteria.html', 'w');
# dados_saida_human = open('/content/drive/My Drive/Data Scientist/human.html', 'w');

dicionario = {};

for i in ['A', 'T', 'C', 'G']:
  for j in ['A', 'T', 'C', 'G']:
    dicionario[i + j] = 0;

for i in range(len(dados_entrada_bacteria) - 1):
  dicionario[dados_entrada_bacteria[i] + dados_entrada_bacteria[i + 1]] += 1;

contador = 1;

for i in dicionario:
  transparencia = str(dicionario[i] / max(dicionario.values()));
  dados_saida_bacteria.write("<div style='width: 100px; height: 100px; border:1px solid #111; float: left; background-color: rgba(0, 255, 0, "+transparencia+"); color = #000'>"+i+"</div>");

  if contador % 4 == 0:
    dados_saida_bacteria.write("<div style='clear: both'><div>");
  
  contador += 1;

dados_saida_bacteria.close();
```
